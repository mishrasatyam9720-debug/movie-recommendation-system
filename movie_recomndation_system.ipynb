```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import difflib
```

```python
df = pd.read_csv("/content/movies (1).csv")
display(df.head())
```

```python
df.isnull().sum()
```

```python
feature_extraction = ['genres', 'keywords', 'tagline', 'cast', 'director']
```

```python
for feature in feature_extraction:
  df[feature] = df[feature].fillna('')
```

```python
categorical_data = df["genres"] + " " + df["keywords"] + " " + df["tagline"] + " " + df["cast"] + " " + df["director"]
display(categorical_data.head())
```

```python
categorical_data.isnull().sum()
```

```python
convertor = TfidfVectorizer()
```

```python
numeric_movie_data=convertor.fit_transform(categorical_data)
print(numeric_movie_data.shape)
```

```python
print(numeric_movie_data)
```

```python
cosine_similarity_matrix = cosine_similarity(numeric_movie_data)

print(cosine_similarity_matrix.shape)
display(cosine_similarity_matrix)
```

```python
movie_name=input("enter the movie name")
```

```python
list_movie=df["title"].tolist()
```

```python
list_movie
```

```python
find_close_match=difflib.get_close_matches(movie_name,list_movie)
closest_movie=find_close_match[0]
closest_movie
```

```python
index_movie=df[df["title"]==closest_movie]["index"].values[0]
index_movie
```

```python
similarity_score=list(enumerate(cosine_similarity_matrix[index_movie]))
```

```python
print(similarity_score)
```

```python
sorted_similar_movies = sorted(similarity_score, key=lambda x: x[1], reverse=True)
sorted_similar_movies
```

```python
print(f'Top 10 Recommendations for {closest_movie}:\n')
i = 1
for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = df[df.index == index]['title'].values[0]
    if i <= 10:
      print(i , " " ,title_from_index)
      i += 1
    else:
      break
```
